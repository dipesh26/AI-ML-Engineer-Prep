### 3. Simulate image loading: Create a 3D array of shape (100, 100, 3) with random pixel values between 0 and 255.

```python
image = np.random.randint(0, 256, size=(100, 100, 3), dtype=np.uint8)

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.title("Random RGB Image")
plt.imshow(image)
plt.grid('off')
plt.show()
```

### 4. Modify that image:
a) Flip horizontally
b) Convert to grayscale using average of RGB

```python
flip_horizontally = np.fliplr(image)
plt.subplot(121)
plt.title("Horizontally Flipped")
plt.imshow(flip_horizontally)

covert_grayscale = np.mean(image, axis=2)
plt.subplot(122)
plt.title("Gray Scale Image")
plt.imshow(covert_grayscale, cmap="gray")
plt.show()
```

### 5. Given this sales data:
sales = np.array([
    ["id", "2021", "2022", "2023"],
    [1, 200, 250, 300],
    [2, 220, 240, 280]
])

Extract only the numeric values, convert to int, and compute average sales per year.

```python
sales = np.array([
    ["id", "2021", "2022", "2023"],
    [1, 200, 250, 300],
    [2, 220, 240, 280]
])

# numeric values, convert to int
numeric_val = sales[1:, 1:].astype(int)
print(f"Numeric Data: \n{numeric_val}\n")

# average sales per year.
years=  sales[0 ,1:]
for i in range(len(years)):
    average_sale = np.mean(numeric_val[:, i])
    print(f"Average Sales in {years[i]}: {average_sale}")
```
