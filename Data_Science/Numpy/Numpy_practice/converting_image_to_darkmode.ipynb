```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

load_image = np.load("numpy-logo.npy")

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(load_image)
plt.title("Numpy Original Logo")
plt.grid(False)


dark_logo = 1 - load_image
        #   or
# dark_logo = np.subtract(1, load_image)

plt.subplot(122)
plt.imshow(dark_logo)
plt.title("Numpy Converted Logo")
plt.grid(False)

plt.show()
```

### Convert to Grayscale

```python
gray_logo = np.mean(load_image, axis=2)
plt.imshow(gray_logo, cmap="gray")
plt.title("Numpy Grayscale Logo")
```

### Flip the Image

```python
flip_vertical = np.flipud(load_image)   # Top-bottom
plt.subplot(121)
plt.imshow(flip_vertical)
plt.title("Vertically Flipped")

plt.subplot(122)
flip_horizontal = np.fliplr(load_image) # Left-right
plt.imshow(flip_horizontal)
plt.title("Horizontal Flipped")
```

### Rotate the Image

```python
rotated_logo = np.rot90(load_image)  # 90 degrees
plt.imshow(rotated_logo)
```
